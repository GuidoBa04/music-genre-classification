```python
import matplotlib.pyplot as plt
import numpy as np
```

```python


# --- DONNES EXTRAITES DES LOGS ---
epochs = range(1, 21)

# Train Loss (Training Loss)
train_loss = [
    1.9236, 1.6040, 1.4622, 1.3330, 1.1929, 
    1.0456, 0.8979, 0.7539, 0.6149, 0.4761, 
    0.3822, 0.3201, 0.2685, 0.2199, 0.1693, 
    0.1754, 0.1359, 0.1491, 0.1429, 0.1181
]

# Train Accuracy
train_acc = [
    25.26, 40.85, 47.48, 53.83, 58.32, 
    64.31, 69.53, 73.40, 78.78, 83.52, 
    87.08, 89.30, 90.30, 92.65, 94.31, 
    94.20, 95.59, 94.92, 95.39, 96.07
]

# Validation Accuracy
val_acc = [
    32.88, 43.12, 48.75, 50.62, 47.75, 
    46.25, 49.00, 44.75, 45.12, 42.50, 
    47.12, 44.75, 42.50, 42.75, 42.25, 
    44.50, 43.50, 43.38, 43.88, 44.88
]

# --- GNRATION DES COURBES ---
plt.figure(figsize=(14, 6))

# Graphique 1 : Evolution de la Loss (Train uniquement ici)
plt.subplot(1, 2, 1)
plt.plot(epochs, train_loss, label='Training Loss', color='tab:blue', marker='o')
plt.title('Descente de Gradient (Training Loss)')
plt.xlabel('poques')
plt.ylabel('Perte (Loss)')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()

# Graphique 2 : Accuracy Train vs Validation (La preuve de l'overfitting)
plt.subplot(1, 2, 2)
plt.plot(epochs, train_acc, label='Training Accuracy', color='tab:blue', marker='o')
plt.plot(epochs, val_acc, label='Validation Accuracy', color='tab:orange', marker='x', linewidth=2)



plt.title('Analyse de l\'Overfitting : Train vs Val')
plt.xlabel('poques')
plt.ylabel('Prcision (%)')
plt.ylim(0, 100)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()

plt.tight_layout()
plt.savefig('baseline_analysis.png', dpi=300)
plt.show()

print("Graphique gnr : 'baseline_analysis.png'")
```
